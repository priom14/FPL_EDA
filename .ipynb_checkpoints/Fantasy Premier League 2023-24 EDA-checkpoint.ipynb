## Comparative study of most goals scored and expected goals scored

```python
most_goals = df[df['goals_scored']>5].sort_values(by = "goals_scored", ascending= False)
most_goals = most_goals[['name','goals_scored']]
most_goals.head(5)
```

```python
most_xg = df[df['name'].isin(most_goals.name)].sort_values(by = 'goals_scored', ascending= False)
most_xg = most_xg[['name', 'goals_scored', 'expected_goals']]
most_xg.head(5)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data = most_xg, x = most_xg.name, y = most_xg.goals_scored, label= 'goals_scored')
sns.lineplot(data = most_xg, x = most_xg.name, y = most_xg.expected_goals, label= 'expected_goals')
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel("Players")
plt.ylabel("Goals Vs. XG")
plt.show()
```

## We will do the comparative study of assistsvs expected assists as well

```python
most_assists = df[df['assists']>5].sort_values(by = "assists", ascending= False)
most_assists = most_assists[['name','assists']]
most_assists.head(5)
```

```python
most_xga = df[df['name'].isin(most_assists.name)].sort_values(by = 'assists', ascending= False)
most_xga = most_xga[['name', 'assists', 'expected_assists']]
most_xga.head(5)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data = most_xga, x = most_xga.name, y = most_xga.assists, label= 'assists')
sns.lineplot(data = most_xga, x = most_xga.name, y = most_xga.expected_assists, label= 'expected_assists')
plt.legend()
plt.xticks(rotation = 90)
plt.xlabel("Players")
plt.ylabel("Assists Vs. Expected Assists")
plt.show()
```

## Top 30 players by Selected by Percentage

```python
plt.figure(figsize=(10,5))
percentdf = df.sort_values("selected_by_percent", ascending = False).head(30)
percentdf_tick = [f"{name} ({team})" for name, team in zip(percentdf["name"], percentdf["team"])]
sns.barplot(data = percentdf, x = percentdf_tick, y = "selected_by_percent", palette= sns.color_palette("flare", len(percentdf)))
plt.xticks(rotation = 90)
plt.xlabel("Players(teams)")
plt.show()
```

## Expected Points for Next match

```python
ep_points = df.sort_values("ep_next", ascending = False).head(20)
ep_ticks = [f"{name} ({team})" for name,team in zip(ep_points["name"], ep_points["team"])]
sns.barplot(data = ep_points, x = ep_ticks, y = "ep_next", palette= sns.color_palette("flare", len(ep_points)))
plt.xticks(rotation = 90)
plt.xlabel("Players(teams)")
plt.show()
```
